## Parámetros

## Main

### Cargar los datos de bronce y visualizar todas sus columnas

```python
limpiar_bronce_df = pd.read_parquet(source_file_path)
```

### Eliminar columna "Unnamed: 0" que no aporta nada y comprobar que se eliminó correctamente

```python
limpiar_bronce_1_df = limpiar_bronce_df.drop('Unnamed: 0', axis=1)
```

```python
limpiar_bronce_1_df.columns
```

```python
limpiar_bronce_1_df.info()
```

### Tratar nulos

```python
limpiar_bronce_1_df[limpiar_bronce_1_df['Financial Status'].isna()== False]['Financial Status']
```

```python
# Defino como voy a tratar los nulos para el caso de la columna "Financial Status", los cuales quiero llamar "Not paid"
# puesto que no se ha realizado el pago todavía.

fill_values = {
    "Financial Status": "Not paid",
}
```

```python
# Aplico la limpieza

limpiar_bronce_2_df = limpiar_bronce_1_df.fillna(fill_values)
```

```python
# Valido que se han corregido correctamente
limpiar_bronce_2_df.info()
```

```python
limpiar_bronce_2_df[limpiar_bronce_2_df['Paid at'].isna()== False]['Paid at']
```

```python
def reemplazar_fechas_erroneas(fila: dict)->None:
    if fila['Paid at'] == None:
        return fila['Created at']
    else:
        return fila['Paid at']
```

```python
limpiar_bronce_paid_at_df = limpiar_bronce_2_df
```

```python
limpiar_bronce_paid_at_df['Paid at'] = limpiar_bronce_paid_at_df.apply(reemplazar_fechas_erroneas, axis=1)
```

```python
limpiar_bronce_paid_at_df['Paid at'].value_counts(dropna=False)

```
