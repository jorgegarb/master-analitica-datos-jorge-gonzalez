### Funciones a usar para limpiar 

```python
"""
Función que convierte los números con formato español o internacional a float


"""
def convertir_decimales_a_float(valor):
    if pd.isna(valor):
        return None
    texto_decimal_normalizado = str(valor).strip().replace('.', '').replace(',', '.')
    try:
        return float(texto_decimal_normalizado)
    except:
        return None
```

```python
def convertir_importes_a_centimos(valor):
    importe = convertir_decimales_a_float(valor)
    if importe is None or pd.isna(importe):
        return pd.NA
    try:
        return int(round(importe * 100))
    except:
        return pd.NA
```

```python
def convertir_fecha_a_utc(valor):
    try:
        timestamp = pd.to_datetime(valor, errors="coerce", utc=True)
        return pd.NaT if pd.isna(timestamp) else timestamp
    except:
        return pd.NaT
```

```python
"""
Marca si un pedido está pagado
"""
def pedido_esta_pagado(fila):
    estado = str(fila.get("Financial Status", "")).strip().lower()
    pagado_por_estado = estado in {"paid", "partially_paid", "authorized"}
    paid_at = convertir_fecha_a_utc(fila.get("Paid at", None))
    return bool(paid_at is not None or pagado_por_estado)
```

```python
def clave_pedido(fila):
    # Clave del pedido (prefiere 'Name')
    if "Name" in fila and fila["Name"] not in (None, "", np.nan):
        return fila["Name"]
    return fila.get("Order ID", None)
```

```python
def codigo_interno_stock_normalizado(fila):
    # SKU o código interno de stock normalizado
    sku = fila.get("Lineitem sku", None)
    if sku in (None, "", np.nan):
        sku = fila.get("Lineitem name", None)
    if sku in (None, "", np.nan):
        return None
    return str(sku).strip().lower()
```

```python
def cantidad_en_linea(fila):
    valor_en_bruto = fila.get("Lineitem quantity", None)
    try:
        if valor_en_bruto in (None, "", np.nan):
            return pd.NA
        normalizar = str(valor_en_bruto).strip().replace(",", ".")
        valor_a_int = int(float(normalizar))
        return pd.array([valor_a_int], dtype="Int64")[0]
    except Exception:
        return pd.NA
```

```python
def importe_total_en_centimos_por_cantidad_en_linea(fila):
    return convertir_importes_a_centimos(fila.get("Lineitem price", None))
```

```python
def total_centimos_linea(fila):
    cantidad = cantidad_en_linea(fila)
    precio_valor = convertir_importes_a_centimos(fila.get("Lineitem price", None))
    if pd.isna(cantidad) or precio_valor is pd.NA:
        return pd.NA
    try:
        precio_centimos = precio_valor[0]
    except Exception:
        precio_centimos = precio_valor
    
    total = int(cantidad) * int(precio_centimos)
    return pd.array([total], dtype="Int64")[0]
```

```python
def es_outlier_iqr(serie):
    la_serie = pd.to_numeric(serie, errors="coerce")
    validos = la_serie.dropna()
    if len(validos) < 4:
        return pd.Series([False] * len(la_serie), index = la_serie.index)
    
    q1 = np.nanpercentile(validos, 25)
    q3 = np.nanpercentile(validos, 75)
    iqr = q3 - q1
    low = q1 - 1.5 * iqr
    high = q3 + 1.5 * iqr

    return (la_serie < low) | (la_serie > high)
```

```python
def convertir_a_booleano_normalizado(valor):
    """
    Normaliza un valor textual/numérico a booleano:
    - Devuelve True si el valor expresa afirmación (sí/si/true/1/yes).
    - Devuelve False si expresa negación (no/false/0/n).
    - Devuelve pd.NA si es nulo o desconocido.
    """
    valor_original = valor

    es_valor_nulo = (valor_original is None or (isinstance(valor_original, float) and pd.isna(valor_original)))
    if es_valor_nulo:
        return pd.NA

    texto_normalizado_minusculas = str(valor_original).strip().lower()

    conjunto_valores_verdaderos = {"sí", "si", "true", "1", "y", "yes"}
    conjunto_valores_falsos = {"no", "false", "0", "n"}

    if texto_normalizado_minusculas in conjunto_valores_verdaderos:
        return True

    if texto_normalizado_minusculas in conjunto_valores_falsos:
        return False

    return pd.NA
```

### Clave de pedido (order_key)

Genero una clave por pedido priorizando "Name" y si no está pues "Order ID" para agrupar líneas y colapsarlo todo a nivel de pedido

```python
# Clave por pedido (prefiere Name y si no, Order ID)
limpiar_bronce_sin_duplicados_df["order_key"] = (limpiar_bronce_sin_duplicados_df.apply(clave_pedido, axis=1))
```

### Creación de columnas auxiliares temporales

Las creo para poder resumir a nivel de pedido (sumas, recuentos y medias) que crean columnas que luego eliminaré

```python
if ("Lineitem price" in limpiar_bronce_sin_duplicados_df.columns) or ("Lineitem quantity" in limpiar_bronce_sin_duplicados_df.columns):
    limpiar_bronce_sin_duplicados_df["__codigo_interno_stock_normalizado"] = (limpiar_bronce_sin_duplicados_df.apply(codigo_interno_stock_normalizado, axis=1))
    limpiar_bronce_sin_duplicados_df["__cantidad_en_linea"] = (limpiar_bronce_sin_duplicados_df.apply(cantidad_en_linea, axis=1))
    limpiar_bronce_sin_duplicados_df["__importe_total_centimos_por_linea"] = (limpiar_bronce_sin_duplicados_df.apply(total_centimos_linea, axis=1))
```

```python
agrupador_por_pedido = limpiar_bronce_sin_duplicados_df.groupby("order_key", dropna=False)
```

```python
suma_cantidades = agrupador_por_pedido["__cantidad_en_linea"].sum(min_count=1).astype("Int64")
skus_unicos = agrupador_por_pedido["__codigo_interno_stock_normalizado"].nunique(dropna=True).astype("Int64")
suma_importe_lineas = agrupador_por_pedido["__importe_total_centimos_por_linea"].sum(min_count=1).astype("Int64")
```

```python
# Precio medio ponderado en céntimos = total_lineas / items_count
denominador_items = suma_cantidades.replace({0: np.nan})
precio_medio_centimos = (suma_importe_lineas / denominador_items).round().astype("Int64")
```

```python
metricas_por_pedido_df = pd.DataFrame({
    "order_key": suma_cantidades.index,
    "items_count": suma_cantidades.values,
    "unique_skus_count": skus_unicos.reindex(suma_cantidades.index).values,
    "avg_item_price_cents": precio_medio_centimos.reindex(suma_cantidades.index).values,
})
```

### Colapso 1 fila = 1 pedido

Ordeno por tiempo y me quedo con la última versión de cada order_key usando drop_duplicates() para limpiar duplicados a nivel pedido

```python
columnas_orden_temporal_para_pedido = []
if "created_at_utc" in limpiar_bronce_sin_duplicados_df.columns:
    columnas_orden_temporal_para_pedido.append("created_at_utc")
if "_BronzeTimestamp" in limpiar_bronce_sin_duplicados_df.columns:
    columnas_orden_temporal_para_pedido.append("_BronzeTimestamp")
```

```python
if columnas_orden_temporal_para_pedido:
    limpiar_bronce_sin_duplicados_df = limpiar_bronce_sin_duplicados_df.sort_values(by=["order_key"] + columnas_orden_temporal_para_pedido)
```

```python
plata_por_pedido_df = limpiar_bronce_sin_duplicados_df.drop_duplicates(subset=["order_key"], keep="last")
plata_por_pedido_df = plata_por_pedido_df.merge(metricas_por_pedido_df, on="order_key", how="left")
```

```python
plata_por_pedido_df["_SilverTimestamp"] = pd.Timestamp.now(tz="UTC")
```

### Seleccion de columnas antes de guardar a plata

```python
columnas_disponibles_previas_a_seleccion = plata_por_pedido_df.columns
columnas_disponibles_previas_a_seleccion
```

```python
lista_columnas_plata_objetivo = [
    "_SilverTimestamp",
    "order_key", "Name", "Risk Level", "Source",
    "created_at_utc", "paid_at_utc", "cancelled_at_utc", "next_payment_due_at_utc",
    "financial_status_bruto", "is_paid",
    "currency",
    "subtotal_cents", "shipping_cents", "taxes_cents", "duties_cents",
    "discount_amount_cents", "order_value_cents",
    "outstanding_balance_cents", "refunded_amount_cents",
    "payment_method", "shipping_method", "discount_code", "shipping_country",
    "accepts_marketing_bool",
    "es_outlier_importe_total", "es_outlier_gasto_envio", "es_outlier_impuestos",
    "es_outlier_aranceles", "es_outlier_descuento", "es_outlier_saldo_pendiente",
    "es_outlier_importe_reembolsado",
    "items_count", "unique_skus_count", "avg_item_price_cents",
]
```

```python
lista_columnas_plata_final = []
for nombre_columna_deseada in lista_columnas_plata_objetivo:
    columna_existe_en_dataframe = (nombre_columna_deseada in plata_por_pedido_df.columns)
    if columna_existe_en_dataframe:
        lista_columnas_plata_final.append(nombre_columna_deseada)
```

```python
# Identificar columnas no seleccionadas y eliminarlas con .drop
columnas_no_seleccionadas = []
for nombre_columna_existente in plata_por_pedido_df.columns:
    forma_parte_de_la_seleccion_final = (nombre_columna_existente in lista_columnas_plata_final)
    if not forma_parte_de_la_seleccion_final:
        columnas_no_seleccionadas.append(nombre_columna_existente)
```

```python
plata_por_pedido_df = plata_por_pedido_df.drop(columns=columnas_no_seleccionadas, errors="ignore")
```

```python
columnas_disponibles_despues_de_drop = list(plata_por_pedido_df.columns)
lista_columnas_en_orden_final = []
for nombre_columna in lista_columnas_plata_final:
    if nombre_columna in columnas_disponibles_despues_de_drop:
        lista_columnas_en_orden_final.append(nombre_columna)
```

```python
plata_por_pedido_df = plata_por_pedido_df[lista_columnas_en_orden_final]
```

```python
columnas_temporales_a_eliminar = []
for nombre_columna_existente in plata_por_pedido_df.columns:
    es_columna_temporal_auxiliar = str(nombre_columna_existente).startswith("__")
    if es_columna_temporal_auxiliar:
        columnas_temporales_a_eliminar.append(nombre_columna_existente)
```

```python
if len(columnas_temporales_a_eliminar) > 0:
    plata_por_pedido_df = plata_por_pedido_df.drop(columns=columnas_temporales_a_eliminar, errors="ignore")
```
