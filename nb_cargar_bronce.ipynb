## nb_cargar_bronce
### Este notebook se usa para cargar el csv "company_exports_bronze.csv" en la carpeta de bronce

## Imports

```python
import os
from datetime import datetime
import pandas as pd 
```

## Parámetros

```python
source_file_path = os.path.join(os.curdir, "company_exports_bronze.csv")
```

```python
bronce_file_name = "bronce.parquet"
bronce_path = os.path.join(os.curdir, "Data_Lake","Bronce")
bronce_file_path = os.path.join(bronce_path, bronce_file_name)
```

## Main

### Cargar el csv a un dataframe

```python
bronce_extract_df = pd.read_csv(source_file_path)
```

```python
# Se comprueba si el fichero se ha cargado correctamente
bronce_extract_df.head()
```

### Añadir el timestamp de carga

```python
bronce_extract_df['_BronzeTimestamp'] = datetime.now()
```

```python
bronce_extract_df.head()
```

### Guardar el DataFrame en parquet

```python
# Creamos el directorio Bronze si no existe
os.makedirs(bronce_path, exist_ok=True)
```

```python
bronce_extract_df.to_parquet(bronce_file_path, index=False)
```
